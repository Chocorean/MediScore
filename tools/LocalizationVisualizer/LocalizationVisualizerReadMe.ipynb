# Localization Visualizer

## Description

This script visualizes select probe and mask images in a specified dataset. The code comes in two separate scripts: get_dims_from_probes.py and get_html_from_probefiles.py.

All csv files passed to the Visualizer must contain headers and must have their rows separated by pipe characters ('|'). Fields and values in the csv should <i>not</i> be enclosed in quotes ( ' or " ) if possible (e.g. entries 'foo', an empty field, and 'bar', in that order, should look like this on the csv: foo||bar, although this won't be checked).

## Command-line Options

Example:

```python
%%bash
python2 get_dims_from_probes.py\
 -t manipulation\
 --refDir ../../data/test_suite/maskScorerTests\
 -r reference/manipulation-image/MFC18-manipulation-image-ref.csv\
 -x indexes/MFC18-manipulation-image-subindex.csv\
 -oR sample_manip_viz_db\
 --filter
```

```python
%%bash
python2 get_html_from_probefiles.py\
 -t manipulation\
 --refDir ../../data/test_suite/maskScorerTests\
 -rx sample_manip_viz_db_refdims.csv\
 -oR sample_manip/sample_manip_viz_db\
 -rr --filter -ow
```

Running get_dims_from_probes.py adds reference information to the index file of desired probes, including the dimensions read from the relevant mask files. Running get_html_from_probefiles.py generates the visual report for these masks, with homepage.html as the homepage for the visual report.

Both files contain the following options:

### Reference

-t --task
 * Two different types of tasks: [manipulation] and [splice]. May include [camera] in future releases.

--refDir
 * Dataset directory path.

-oR --outRoot
 * Directory root plus prefix to save outputs. Will save as [directory]/[prefix]_refdims.csv, separated by pipes "|".

### get_dims_from_probes.py
The following options are specific to get_dims_from_probes.py:

### Input Options

-r --inRef
 * Reference csv file name relative to --refDir
 
-x --inIndex
 * Task Index csv file name relative to --refDir. The index file can be customized by the user depending on the probes to display.

### Other Options

--filter
 * Filter out rows with empty masks (since they would not apply to dimensionality discrepancy anyway).

### get_html_from_probefiles.py
The following options are specific to get_html_from_probefiles.py:

### Input Options

-rx --refIndex
 * The reference joined with the index generated by get_dims_from_probes.py

-soR --sourceOutRoot
 * Original output root of the HTML pages. Primarily important when generating subsets of the HTML pages when the existing ones have been generated. Default: argument passed to --outRoot.

### Output Options

-N
 * The number of probes per homepage partition. 0 will output all the rows to one homepage; numbers greater than 0 will be output to separate directories suffixed by '_k', where k is the number of the partition in order of JournalName.
 
-hp --homepage_only
 * Generate the homepage only. Should be used only if the individual pages are all generated. If used with -N, will also generate symbolic links to the generated probes in the original directory.

--compress
 * Save images scaled to a percentile factor. For example, passing a factor of 20 will scale the image to 20% of its original dimensions. If used with -hp and , will override -hp. Default: 100.

-rr --reduce_redundancy
 * Do not generate a probe mask image or donor mask image for the homepage display (due to existence of overlaid images).

--filter
 * Filter out the rows for which no badness has been detected.
 
-ow --overwrite
 * Overwrite existing images if they exist.
